```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset)
dataset=dataset.replace({True:1, False:0})
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'sex_female','smoker_no','smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly', C=100)
regressor.fit(X_train, y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
# Test set predicted method
y_pred=regressor.predict(X_test)
y_pred
```

```python
# Evovation metrics 
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
# Saved best model
import pickle
filename="finalized_model_svmr.sav"
```

```python
pickle.dump(regressor,open(filename, 'wb'))
```

```python
# Loading best model and verify
loaded_model=pickle.load(open("finalized_model_svmr.sav", 'rb'))
result=loaded_model.predict([[25,26.5,2,1,0,1,0]])
result
```
